```python
#Univariate Analysis - Statistical Non Visual Analysis

discrete_df = df.select_dtypes(include=['object'])

numerical_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
def discrete_univariate_analysis(discrete_data):
    for col_name in discrete_data:
        print("*"*10, col_name, "*"*10)
        print(discrete_data[col_name].agg(['count', 'nunique', 'unique']))
        print('Value Counts: \n', discrete_data[col_name].value_counts())
        print()
```

```python
discrete_univariate_analysis(discrete_df)
```

```python
def numerical_univariate_analysis(numerical_data):
  for col_name in numerical_data:
    print("*"*10, col_name, "*"*10)
    print(numerical_data[col_name].agg(['min', 'max', 'mean', 'median', 'std']))
    print()
```

```python
numerical_univariate_analysis(numerical_df)
```

```python
#Bivarite Analysis ON cat vs num
top_10_specializations_by_gender = df.groupby(['Gender',
'Specialization'])['Salary'].mean().unstack()
top_10_specializations_by_gender =top_10_specializations_by_gender.apply(lambda x: x.nlargest(10),axis=1)
plt.figure(figsize=(12, 6))
top_10_specializations_by_gender.plot(kind='bar', figsize=(10, 6))
plt.title('Top 10 Specializations with Highest Average Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Salary')
plt.legend(title='Specialization',loc='center left',
bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
top_10_specializations = df.groupby('Specialization')['Salary'].mean().nlargest(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_specializations.values,
y=top_10_specializations.index)
plt.title('Top 10 Specializations with Highest Average Salary')
plt.xlabel('Average Salary')
plt.ylabel('Specialization')
plt.show()
```
